```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('dados.csv', sep=';')
data
```

```python
sb.distplot(data.PIB, bins=3, axlabel='PIB').set_title('PIB')
```

```python
sb.distplot(data.VALOREMPENHO, bins=6, axlabel='Valor Empenho').set_title('Valor Empenho')
```

```python
agrupado = data.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO', y='PIB', color='gray')
```

```python
agrupado = data.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```
