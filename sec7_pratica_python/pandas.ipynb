```python
# séries, única coluna

# pode ser criada a partir de listas, array do numpy ou coluna de dataframe
s1 = pd.Series([2,4,2,5,6,4,6,13])
print(s1)
```

```python
# renomear de forma não persistida 
dados.rename(columns={'duration':'duracao','purpose':'proposta'})
```

```python
# renomear de forma persistida
dados.rename(columns={'duration':'duracao','purpose':'proposta'}, inplace=True)
dados.head(1)
```

```python
# excluir coluna
dados.drop('checking_status', axis=1, inplace=True)
dados.head(1)
```

```python
# verificar dados nulos
dados.isnull()
```

```python
# verificar dados nulos de uma grande quantidade de dados
dados.isnull().sum()
```

```python
# preenchimento de dados faltantes
dados['duracao'].fillna(0, inplace=True)
```
