```python
# Histograma

# Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# carregamento da base de dados
data = pd.read_csv('trees.csv')
data.shape
```

```python
# dados
data
```

```python
# criacao do histograma, considerando somente o segundo atributo dos dados e com duas divisoes

# a variavel 'h' armazena as faixas de valores de Height
h = np.histogram(data.iloc[:,1], bins=6)
h
```

```python
# visualizacao do histograma com 6 divisoes (bins)
plt.hist(data.iloc[:,1], bins=6)
plt.title('Arvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```
