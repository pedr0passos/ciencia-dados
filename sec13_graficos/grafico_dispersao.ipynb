```python
# importacoes
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
# carregamento dos dados 
data = pd.read_csv('trees.csv')
data
```

```python
# grafico de dispersao considerando o volume e a circunferencia 
plt.scatter(data.Girth, data.Volume, color='blue', facecolors='none', marker='*')
plt.title('Arvores')
plt.xlabel('Volume')
plt.ylabel('Circunferencia')
```

```python
# grafico de linha considerando o volume e o atributo 'girth'
plt.plot(data.Girth, data.Volume)
plt.title('Arvores')
plt.xlabel('Volume')
plt.ylabel('Circunferencia')
```

```python
# grafico de dispersao com 'afastamento' dos dados (jitter)

#fit_reg = linha de tendencia
sb.regplot(data.Girth, data.Volume, data = data, x_jitter = 0.3, fit_reg=False)
```
