```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('co2.csv')
data
```

```python
# grafico de dispersao utilizando os atributos conc e uptake, agrupamento pelo type
sb.scatterplot(data.conc, data.uptake, hue = data.Type)
```

```python
# selecao de registros especificos da base de dados ( quebec e mississipi )
q = data.loc[data['Type'] == 'Quebec']
m = data.loc[data['Type'] == 'Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
sb.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sb.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
# refrigerado e nÃ£o refrigerado
ch = data.loc[data['Treatment'] == 'chilled']
nc = data.loc[data['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
sb.scatterplot(ch.conc, ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
sb.scatterplot(nc.conc, nc.uptake).set_title('Non Chilled')
plt.tight_layout()
```
