```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# criacao de uma variavel com dados em uma distribuicao normal com a funcao rvs (100 elementos)
dados = norm.rvs(size = 100)
dados
```

```python
# histograma
plt.hist(dados, bins=20)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
# execucao do teste de Shapira
# segundo argumento é o valor de p, não há como rejeitar a hipotese nula
stats.shapiro(dados)
```

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
# histograma
plt.hist(dados2, bins=20)
plt.title('Dados2')
```

```python
# geracao de graficos para verificar se a distribuicao é normal
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot = ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
